```python
import pandas as pd

df = pd.read_csv('netflix_titles.csv')
df
```

```python
df.isnull().sum()
```

```python
# Number of movies vs TV shows
df['type'].value_counts().plot(kind='bar', title='Content Type')

```

```python
# Top 10 countries with most content
df['country'].value_counts().head(10).plot(kind='barh', title='Top Countries')

```

```python
# Most popular genres
df['listed_in'].str.split(',').explode().value_counts().head(10).plot(kind='bar', title='Popular Genres')
```
